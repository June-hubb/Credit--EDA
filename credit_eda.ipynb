```python
def age(x):
    if x>=20 and x<30:
        return '20-30'
    elif x>=30 and x<40:
        return '30-40'
    elif x>=40 and x<50:
        return '40-50'
    elif x>=50 and x<60:
        return '50-60'
    elif x>=60 and x<70:
        return '60-70'
```

```python
app['AGE_GROUP']= app['AGE'].apply(age)
```

```python
app['EXCESS_AFTER_LOAN'] = app['AMT_INCOME_TOTAL']-app['AMT_ANNUITY']
```

```python
def excess(x):
    if x <=0:
        return 'Less than or equal to 0'
    elif x>0 and x<1000:
        return '0-1000'
    elif x>=1000 and x<10000:
        return '1000-10000'
    elif x>=10000 and x<100000:
        return '10000-100000'
    elif x>=100000 and x<1000000:
        return '100000-1000000'
    elif x>=1000000:
        return 'More than 1000000'
```

```python
app['EXCESS_AFTER_LOAN']=app['EXCESS_AFTER_LOAN'].apply(excess)
plt.figure(figsize=(8,4))
g=sns.catplot(x="EXCESS_AFTER_LOAN", kind="count", palette="ch:.25", data=app)
g.set_xticklabels(rotation=90)
plt.show()
```
